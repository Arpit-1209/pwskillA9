1. Basic Convolution Operation Using a Filter and a 5x5 Image

```python
import numpy as np

# Define a 5x5 image (matrix)
image = np.array([
    [1, 2, 3, 0, 1],
    [0, 1, 2, 3, 1],
    [3, 0, 1, 2, 0],
    [2, 3, 0, 1, 2],
    [1, 2, 3, 0, 1]
])

# Define a 3x3 filter (kernel)
filter = np.array([
    [1, 0, -1],
    [1, 0, -1],
    [1, 0, -1]
])

# Perform convolution (valid padding)
output_shape = (image.shape[0] - filter.shape[0] + 1, image.shape[1] - filter.shape[1] + 1)
output = np.zeros(output_shape)

for i in range(output_shape[0]):
    for j in range(output_shape[1]):
        region = image[i:i+3, j:j+3]
        output[i, j] = np.sum(region * filter)

print("Convolved Output:\n", output)
```

2. Max Pooling on a 4x4 Feature Map with a 2x2 Window


```python
# Define a 4x4 feature map
feature_map = np.array([
    [1, 3, 2, 1],
    [4, 6, 5, 2],
    [7, 3, 8, 6],
    [2, 4, 3, 7]
])

# Max pooling with a 2x2 window
pooled_output = np.zeros((2, 2))

for i in range(0, 4, 2):
    for j in range(0, 4, 2):
        pooled_output[i//2, j//2] = np.max(feature_map[i:i+2, j:j+2])

print("Max Pooled Output:\n", pooled_output)
```

3. ReLU Activation Function on a Feature Map

```python
# Define a feature map
feature_map = np.array([
    [-1, 3, -2, 1],
    [4, -6, 5, -2],
    [7, -3, 8, 6],
    [-2, 4, -3, 7]
])

# Apply ReLU activation
relu_output = np.maximum(0, feature_map)

print("ReLU Output:\n", relu_output)
```

5. Generate Synthetic Dataset and Train CNN

```python
from sklearn.model_selection import train_test_split

# Generate synthetic data
X = np.random.random((1000, 28, 28, 1))  # 1000 random images
y = np.random.randint(0, 2, 1000)  # Binary labels

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Define and train the model
model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(8, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
history = model.fit(X_train, y_train, epochs=5, validation_data=(X_test, y_test))
```

11. Plot Accuracy and Loss

```python
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense

# Sample dataset
import numpy as np
X_train = np.random.rand(100, 4)  # 100 samples, 4 features
y_train = np.random.randint(0, 2, 100)  # Binary target (0 or 1)
X_val = np.random.rand(20, 4)  # 20 validation samples
y_val = np.random.randint(0, 2, 20)  # Binary target for validation

# Create the model
model = Sequential([
    Dense(10, input_dim=4, activation='relu'),
    Dense(1, activation='sigmoid')
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=10, batch_size=8)

# Plot accuracy
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# Plot loss
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
